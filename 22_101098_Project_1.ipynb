```python
df_filtered = df[df['Have you participated in any internships during your college years?'] == 'Yes']

df_filtered["If yes, how many internships have you completed?"] = pd.to_numeric(df_filtered["If yes, how many internships have you completed?"], errors='coerce')

df_filtered = df_filtered.dropna(subset=["If yes, how many internships have you completed?"])

plt.figure(figsize=(8, 6))
plt.hist(df_filtered["If yes, how many internships have you completed?"], bins=5, edgecolor='black')
plt.xlabel("Number of Internships Completed")
plt.ylabel("Number of Students")
plt.title("Distribution of Internship Experiences")
plt.grid(axis='y')
plt.show()
```

```python
column1 = "If yes, how many internships have you completed?"
column2 = "How would you rate the overall impact of your internships on your acquisition of practical skills relevant to your desired career field? (Scale: Minimal impact (1) - Significant impact (5))"

df[column1] = pd.to_numeric(df[column1], errors='coerce')
df[column2] = pd.to_numeric(df[column2], errors='coerce')

df.dropna(subset=[column1, column2], inplace=True)
correlation = df[column1].corr(df[column2])

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df[column1], df[column2])
plt.xlabel(column1)
plt.ylabel(column2)
plt.title(f"Correlation between {column1} and {column2}: {correlation:.2f}")
plt.grid(True)
plt.show()
```
